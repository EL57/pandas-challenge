* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results

```python
#Grouping the purchase data by non null unique values
grouped_gender = purchase_data.groupby('Gender').nunique()
#print(grouped_gender.head(5))

#Creating the column for Purchase count by counting the purchase IDs
grouped_gender['Purchase Count'] = purchase_data.groupby('Gender')['Purchase ID'].count()

grouped_gender['Purchase Count']

#Creating the column for the average price 
grouped_gender['Average Purchase Price'] = purchase_data.groupby('Gender')['Price'].mean()

grouped_gender['Average Purchase Price']

#Creating the column for the total purchase value
grouped_gender['Total Purchase Value'] = purchase_data.groupby('Gender')['Price'].sum()

grouped_gender['Total Purchase Value']

#Creating the average total purchase per person column by dividing the total by the unique people.
grouped_gender['Avg Total Purchase per Person'] = round(grouped_gender['Total Purchase Value']/grouped_gender['SN'], 2)

grouped_gender['Avg Total Purchase per Person']
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Finding the largest and smallest age
purchase_data['Age'].max()
purchase_data['Age'].min()
```

```python
#Creating bins for the age groups
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
#Labels for the age ranges
group_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
#applying the bins to the purchase data
pd.cut(purchase_data['Age'], bins, labels=group_labels).head()
```

```python
#Adding "Age Group" bins to the purchase data
purchase_data["Age Group"] = pd.cut(purchase_data['Age'], bins, labels=group_labels)
purchase_data.head()
```

```python
# Create a GroupBy object based upon "Age Group"

grouped_ages = purchase_data.groupby('Age Group').nunique()

#Create a total of unique players variable for the perrcentage 
total_count = grouped_ages['SN'].sum()

#Create the percentage column
grouped_ages['Percentage of Players'] = round((grouped_ages['SN'] / total_count) * 100, 2)

#I wanted the column in the table to say Total Count and not SN. Theis actually worked unline .rename...should research
grouped_ages['Total Count'] = grouped_ages['SN']

#Displaying table
grouped_ages[['Total Count','Percentage of Players']]
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Grouping the purchase data by unique names (SN)
grouped_names = purchase_data.groupby('SN').nunique()

#Getting the average purchase of every player
grouped_names['Purchases'] = round(purchase_data.groupby('SN')['Price'].mean(), 2)

#Getting the purchase counts for every player
grouped_names['Purchase Count'] = purchase_data.groupby('SN')['Purchase ID'].count()

#Getting the total purchases 
grouped_names['Total Purchase Value'] = purchase_data.groupby('SN')['Price'].sum()

#Creating a variable for the data to sort
grp_nm_dis = grouped_names[['Purchase Count','Purchases','Total Purchase Value']]

#Sorting data on Total Purchase Value
grp_nm_dis.sort_values("Total Purchase Value", ascending=False).head()


#print(grouped_names['Total Purchase Value'].max())
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#purchase_data.head(2)
#purchase_data[['Item ID','Item Name','Price']]

#Group purchase data by unique item id and item name
grouped_items = purchase_data.groupby(['Item ID', 'Item Name']).nunique()

#Getting the purchase count by counting the purchase IDs
grouped_items['Purchase Count'] = purchase_data.groupby(['Item ID', 'Item Name'])['Purchase ID'].count()

#Getting the total purchase value by summing the price
grouped_items['Total Purchase Value'] = purchase_data.groupby(['Item ID', 'Item Name'])['Price'].sum()

#Getting the item price
grouped_items['Item Price'] = purchase_data.groupby(['Item ID', 'Item Name'])['Price'].mean()

#Displaying what I have
grouped_items[['Purchase Count','Item Price','Total Purchase Value']]

#Creating a variable for the data to sort
grp_it_dis = grouped_items[['Purchase Count','Item Price','Total Purchase Value']]

#grp_it_dis
#Sorting data on Purchase Count descending and showing the first 5 rows
grp_it_dis.sort_values("Purchase Count", ascending=False).head()

#print(grouped_items)
```
